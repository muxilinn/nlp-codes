```python
import nltk
from tokenizers import ByteLevelBPETokenizer
from nltk.corpus import brown

# download NLTK corpora
nltk.download('brown')
nltk.download('punkt')
```

```python
# 1. Prepare training data from NLTK Brown corpus
train_text = '\n'.join(brown.words(categories='news'))  # 提取新闻类文本
print(f"# Charcters: {len(train_text)}")
print(f"Ex.: {train_text[:100]}")

with open('bpe_train.txt', 'w', encoding='utf-8') as f:
    f.write(train_text)
```

```python
# 2. Initalize BPE tokenizer and train
tokenizer = ByteLevelBPETokenizer()
tokenizer.train(
    files=['bpe_train.txt'],
    vocab_size=5000,
    min_frequency=2,
    special_tokens=['<pad>', '<bos>', '<eos>', '<unk>']
)

# 3. save and load the tokenizer
tokenizer.save_model('.', 'bpe_tokenizer')
tokenizer = ByteLevelBPETokenizer('bpe_tokenizer-vocab.json', 'bpe_tokenizer-merges.txt')

# 4. Test by BPE
text = "Natural language processing is fun! Let's try BPE tokenization."
sentences = nltk.sent_tokenize(text)

for sent in sentences:
    print("Text: ", sent)
    encoded = tokenizer.encode(sent)
    print("BPE subwords:", encoded.tokens)
    print("ID:", encoded.ids)
    print("---")
```
